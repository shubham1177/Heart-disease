# Preprocessing

```python
df.dtypes
```

```python
df.columns
```

converting the Data into Categorical to Numeric

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
df['ChestPainType']=le.fit_transform(df['ChestPainType'])
df['RestingECG']=le.fit_transform(df['RestingECG'])
df['ExerciseAngina']=le.fit_transform(df['ExerciseAngina'])
df['ST_Slope']=le.fit_transform(df['ST_Slope'])
```

```python
df.head()
# checking the data again
```

```python
df.dtypes
# checking dtypes as well
```

spliting the data into two parts

```python
x = df.drop('HeartDisease',axis=1)
```

```python
y = df['HeartDisease']
```

last step of preprocessing spliting the data into Train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123,stratify=y)
```

# Importing Algorithm



1. Data is binary classification problem so first we import Logistic regression
2. As Data is small nad linear so go with NaiveBayes
3. There are 7 Catergorical columns so we will be implementing Decision Tree


Importing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

Importing NaiveBayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
y_pred =nb.predict(x_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

Importing DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred =dt.predict(x_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

1. LogisticRegression gives 82% accuracy for the model
2. Naive bayes gives 85% accuracy for the model
3. DecisionTree gives 77% accuracy for the model
